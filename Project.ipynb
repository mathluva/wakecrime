```python
#bulk export csv files, only 6 months allowed at
#a time, http://www.wakegov.com/ccbi/publicservices/Pages/default.aspx

crime1=pd.read_csv("1-Jan_May31_2015.csv")
crime2=pd.read_csv("2-Jun_Nov30_2015.csv")
crime3=pd.read_csv("3-Dec2015_May31_2016.csv")
crime4=pd.read_csv("4-20160601_20161130.csv")
crime5=pd.read_csv("5-20161201_20170531.csv")
crime6=pd.read_csv("6-20170601_20171130.csv")
```

```python
crime7=pd.read_csv("7-20171201_20180531.csv")
crime8=pd.read_csv("8-20180601_20181130.csv")
crime9=pd.read_csv("9-20181201_20190531.csv")
crime10=pd.read_csv("10-20190601_20191130.csv")
crime11=pd.read_csv("11-20191201_20200530.csv")
crime12=pd.read_csv("12_20200601_20201024.csv")
```
